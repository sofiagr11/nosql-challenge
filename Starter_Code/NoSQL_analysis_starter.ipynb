### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python

# Get the geocode of the "Penang Flavours" restaurant
penang_flavours = establishments_collection.find_one({"BusinessName": "Penang Flavours"})
penang_flavours_geocode = penang_flavours.get("geocode", {})

# Define the latitude and longitude range
degree = 0.01
latitude_range = {"$gte": penang_flavours_geocode["latitude"] - degree, "$lte": penang_flavours_geocode["latitude"] + degree}
longitude_range = {"$gte": penang_flavours_geocode["longitude"] - degree, "$lte": penang_flavours_geocode["longitude"] + degree}

# Define the query to find establishments with RatingValue of 5 within the specified latitude and longitude range
query6 = {
    "RatingValue": 5,
    "geocode.latitude": latitude_range,
    "geocode.longitude": longitude_range
}

# Define the projection to only include necessary fields
projection = {"BusinessName": 1, "scores.Hygiene": 1, "geocode": 1}

# Sort the establishments by hygiene score in ascending order
sort = [("scores.Hygiene", pymongo.ASCENDING)]

# Execute the query and sort the results
top_establishments = establishments_collection.find(query6, projection).sort(sort).limit(5)

# Display the results

for establishments_collection in top_establishments:
    print(establishments_collection)
```

```python
# :((
# Convert result to Pandas DataFrame
data3 = list(top_establishments)

# Convert the result to a DataFrame
df3 = pd.DataFrame(data3)

# Display the DataFrame
print(df3)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Define the aggregation pipeline
pipeline = [
{"$match": {"scores.Hygiene": 0}},  # Match establishments with a hygiene score of 0
{"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},  # Group by Local Authority and count establishments
{"$sort": {"count": -1}},  # Sort the results from highest to lowest]
]
# Execute the aggregation query
result_4 = establishments_collection.aggregate(pipeline)
result_4_2 = list(establishments_collection.aggregate(pipeline))
```

```python
# Print the number of documents in the result
num_docs = len(result_4_2)
print("Number of documents:", num_docs)
```
